# 4. The dataset is not in clean form. Write a function that takes pandas dataframe as an input and replace these characters i-e ( "\t", " ", "?") with "nan" and again check unique values of Diabetes Mellitus(dm)

```python
ckdf.loc[(ckdf.dm !='yes') & (ckdf.dm !='no') &(ckdf.dm !='nan')]  = np.nan  #using nupmy.nan to fill value other than yes or no
```

```python
ckdf.groupby('dm').dm.count() #Proving there is no "?yes" occurance as there was before
```
