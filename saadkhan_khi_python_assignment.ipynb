```python
import pandas as pd
import os
import matplotlib.pyplot as plt

def clean_dataframe(df):
    func = lambda x: str(x).strip()
    df=df.applymap(func)    
    df=df.replace("?",float('nan'))
    df=df.replace("nan",float('nan'))
    return df
def print_unique_values(df):
    size=len(df.columns)
    for x in range(0,size):
        print(str(df.columns[x])+" "+str(df.iloc[:,x].unique()))


#Q1
os.chdir("C:/Users/Saad.Khan/Documents")
dataset=pd.read_csv("Assignment Material/Assignment Material/Dataset/chronic_kidney_disease_updated.csv")
#print_unique_values(dataset)
print("Q1: Load the dataset to pandas dataframe and print names of columns\n\n")
print(dataset.columns)
print("\n\n")


#Q2
print("Q2: Display first 5 rows in the dataset.\n\n")
print(dataset.loc[0:4,:])
print("\n\n")

#Q3
print("Q3: Check if dataset is in cleaned form else clean it. Check Diabetes Mellitus(dm) variable and print what type of values it has.\n\n")
print(dataset.dm.unique())
print("\n\n")

#Q4
dataset=clean_dataframe(dataset)
print("Q4: The dataset is not in clean form. Write a function that takes pandas dataframe as an input, remove these characters i-e ( '\t', ' ') from the string value and replace with 'nan' if the string value has only unicode characters i-e ('?'). The function should apply this transformation to all columns in dataframe and for cross check display unique values of Diabetes Mellitus(dm).\n\n")
print(dataset.dm.unique())
print("\n\n")

#Q5
print("Q5: Change datatype to Numerical of following columns in Dataframe. columns 'age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc'\n\n")

dataset[['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']]=dataset[['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']].astype(float)
print("Type changed.\n\n")

#Q6
print("Q6: Count the number of normal or abnormal red blood cells people having Chronic Kidney Disease(CKD).\n\n")
print("Abnormal")
print(dataset.loc[(dataset["rbc"] == "abnormal") & (dataset["class"] == "ckd"),"rbc"].count())
print("Normal")
print(dataset.loc[(dataset["rbc"] == "normal") & (dataset["class"] == "ckd"),"rbc"].count())
dataset[dataset["class"]=="ckd"].rbc.value_counts().plot(kind='bar')
plt.show()

print("\n\n")

#Q7
print("Q7: Display maximum blood pressure having Chronic Kidney Disease(CKD).\n\n")
print(dataset.loc[:,"bp"].max())
print("\n\n")

#Q8
print("Q8: Create a csv file of cleaned data in dataframe with name 'clean_chronic_kidney_disease.csv'\n\n")
dataset.to_csv("clean_chronic_kidney_disease.csv", sep=',', encoding='utf-8',index=False)
clean_dataset=pd.read_csv("clean_chronic_kidney_disease.csv")
#print_unique_values(clean_dataset)
print("File successfully created\n\n")
```
